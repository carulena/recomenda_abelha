```python
def extrairValorEmLote(df, tiff_path,nome_coluna_saida,  coluna_lat='Latitude', coluna_lon='Longitude'):
    with rasterio.open(tiff_path) as src:
        banda = src.read(1)
        transform = src.transform

        def extrair_valor(row):
            try:
                row_idx, col_idx = rowcol(transform, row[coluna_lon], row[coluna_lat])
                return banda[row_idx, col_idx]
            except:
                return None

        df[nome_coluna_saida] = df.apply(extrair_valor, axis=1)
    
    return df
```

```python
df_total = extrairValorEmLote(df_total, '../Dados/Clima/Preciptação/historico_media.tif','Precipitação')
df_total = extrairValorEmLote(df_total, '../Dados/Clima/Temperatura Máxima/historico_media.tif', 'TempMax')
df_total = extrairValorEmLote(df_total, '../Dados/Clima/Temperatura Mínima/historico_media.tif', 'TempMin')
df_total['TempMedia'] = df_total[['TempMax', 'TempMin']].mean(axis=1)
df_total
```

```python

import pandas as pd
from scipy.spatial import cKDTree

df_total['especie_abelha'] = [[] for _ in range(len(df_total))]


tree = cKDTree(df_ocorrencias[['Latitude', 'Longitude']].values)

tolerancia = 0.01

def especies_proximas(lat, lon):
    ponto = [lat, lon]
    indices = tree.query_ball_point(ponto, r=tolerancia)
    if indices:
        return df_ocorrencias.iloc[indices]['Especie'].unique().tolist()
    else:
        return []

df_total['especie_abelha'] = df_total.apply(
    lambda row: especies_proximas(row['Latitude'], row['Longitude']), axis=1
)

df_total['ocorrencia'] = df_total['especie_abelha'].apply(lambda x: len(x) > 0)
```
